# wsl_data.csv

```python
#wsl_data
df_wsl = pd.read_csv('https://raw.githubusercontent.com/nikojim/coderhouse/main/wls_data.csv')
df_wsl.head()
```

```python
#describe + info
df_wsl.describe()
```

```python
df_wsl.info()
```

```python
#para poder realizar cálculos sobre ciertas variables se rellenan los valores nulos a cero
df_wsl.fillna(value=0)
```

```python
#Para trabajar mejor con el dataset voy a quedarme solo con las columnas más importantes
df_wsl.drop(df_wsl.columns[[6,8]], axis=1, inplace=True)
```

```python
df_wsl.head()
```

```python
#obtegno el rango de años sobre los que hay información
print('El dataframe contiene información desde {} hasta {}'.format(df_wsl.year.min(),df_wsl.year.max()))
```


**Media, mínimo y máximo de altura y peso de surfistas agrupados por país**

```python
df_aux = df_wsl.groupby(('athlete_country')).agg({'athlete_height': ['mean', 'min', 'max'],'athlete_weight': ['mean', 'min', 'max']})
df_aux.columns = ['athlete_height_mean', 'athlete_height_min', 'athlete_height_max','athlete_weight_mean', 'athlete_weight_min', 'athlete_weight_max']
df_aux = df_aux.reset_index()
df_aux.dropna()
```

**Top 10 surfistas que han ganado más eventos**

Nota: cuándo se gana un evento se le otorgan 10.000 puntos al ganador, por eso es la condición de filtro.

```python
#quiero saber los 10 surfistas que han ganado más eventos
df_aux = df_wsl[df_wsl.points_earned == 10000]
df_aux = df_aux[['athlete_name']]
df_aux.value_counts().head(10).reset_index()
```

**Top 10 surfistas que han ganado más puntos**

```python
#quiero saber los 10 surfistas que han ganado más puntos
df_aux = df_wsl[['athlete_name', 'points_earned']]
df_aux.groupby('athlete_name').sum().nlargest(10,'points_earned').reset_index()
```

**Todos los ganadores de los eventos del año 2022**

```python
#todos los ganadores de los eventos del año 2022
df_wsl[(df_wsl.points_earned == 10000) & (df_wsl.year==2022)].sort_values(by=['event_order']).reset_index()
```

**Todos los surfistas que ganaron Bells  ordenado por año**

```python

#todos los surfistas que ganaron Bells  ordenado por año
df_wsl[(df_wsl.event_name.str.contains('Bells')) & (df_wsl.points_earned == 10000)].sort_values(by='year').reset_index()
```

**Todos los campeones del circuito por año**

```python
#todos los campeones del circuito por año
df_aux = df_wsl.groupby(['year', 'athlete_name']).agg({'points_earned': 'sum'})
df_aux.columns = ['points_earned_sum']
df_aux = df_aux.reset_index()
df_aux.groupby(["year"]).apply(lambda x: x.nlargest(1, "points_earned_sum")).reset_index(drop=True)
```

**Suma de los top 5 puntos de los surfistas por año agrupados por país**

```python
#suma de puntos por año por país
df_aux = df_wsl.groupby(['year', 'athlete_country']).agg({'points_earned': 'sum'})
df_aux.columns = ['points_earned_sum']
df_aux = df_aux[df_aux.groupby(['year'])['points_earned_sum'].rank(method='first', ascending=False) <= 5]
df_aux = df_aux.reset_index()
df_aux
```

**Gráfico del total de puntos de surfistas agrupados por país en todos los años**

Se puede inferir que el puntaje de los surfistas australianos ha decrecido con los años y los surfistas brasileños han progresado mucho.

Los surfistas estadounidenses luego de una baja participación han mantenido una puntuación bastante regular.

Los surfistas franceses y hawaianos han mantenido una puntuación regular.

Japón ha aparecido en los puntajes top debido a la performance de Kanoa Igarashi.

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_aux, x='year', y='points_earned_sum', hue='athlete_country', palette = 'coolwarm')
```

**gráfico similar al anterior pero con líneas de regresión**

```python
sns.lmplot(x='year',y='points_earned_sum',data=df_aux,hue='athlete_country',palette='coolwarm')
```

```python
# top 5 surfistas puntuados en los últimos 3 años (en 2020 no hubo competencia debido a covid-19)
df_aux = df_wsl[df_wsl.year >= 2019].groupby(['year', 'athlete_country','athlete_name']).agg({'points_earned': 'sum'})
df_aux.columns = ['points_earned_sum']
df_aux = df_aux[df_aux.groupby(['year'])['points_earned_sum'].rank(method='first', ascending=False) <= 5]
df_aux = df_aux.reset_index()
df_aux.sort_values(['year','points_earned_sum'],ascending = False)
```

**Cantidad de partidas por año y agregación de resultados**

```python
#Cantidad de partidas por año y agregación de resultados
df_aux = df_games.groupby(['Year', 'Result']).agg({'Result': 'count'})
df_aux.columns = ['Result_count']
df_aux = df_aux.reset_index()
df_pivot = df_aux.pivot(index='Year',columns='Result',values='Result_count')
df_pivot['Total Partidas'] = df_pivot.sum(axis=1)
df_pivot



```

```python
df_aux = df_games.groupby(['Year', 'Result']).agg({'Result': 'count'})
df_aux.columns = ['Result_count']
df_aux = df_aux.reset_index()
df_aux
```

**Gráficos de partidas por años y sus resultados**

Se puede inferir que con el avance de los años hay una tendencia mayor a las tablas (draw) y por lo tanto han disminuído tanto las victorias con blancas y con negras.

```python
plt.figure(figsize=(20,6))
df_aux = df_aux [df_aux.Year != '1985']
sns.lineplot(data=df_aux, x='Year', y='Result_count', hue='Result',palette='coolwarm')
```

**Desglose y gráfico por top 20 de aperturas jugadas**

La apertura que ha generado más victorias de blancas es la C52 - Gambito Evans.

La apertura que ha generado más tablas es la D58 - Gambito de dama declinado
Tartakower

La apertura que ha generado más victorias de negras es la D52 - Gambito de dama declinado.


```python
#Agrupación por top 20 de aperturas y número de partidas
df_aux = df_games.groupby(['ECO','Opening','Result']).agg({'Result': 'count'})
df_aux.columns = ['Result_count']
df_aux = df_aux.reset_index()
df_pivot = df_aux.pivot(index=['ECO','Opening'],columns='Result',values='Result_count')
df_pivot['Total Partidas'] = df_pivot.sum(axis=1)
df_top_20 = df_pivot.fillna(value=0).sort_values('Total Partidas',ascending = False).head(20)
df_top_20 = df_top_20.reset_index()
df_top_20





```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=df_top_20, x='ECO',y='Total Partidas',palette = 'coolwarm')
```

```python
df_top_20_simple = df_top_20[['ECO','White','Black','Draw']]
df_top_20_simple.reset_index()
df_top_20_simple.set_index('ECO').plot(kind='bar', stacked=True, use_index=True)
```

+0.4 to 0.6	 | Moderate + association

```python
#correlaciones
df_movie_simple[['imdb_score','movie_facebook_likes','gross','budget','director_facebook_likes','actor_1_facebook_likes','cast_total_facebook_likes','duration']].corr() 
```

Se nota también una correlación de **0.254709** entre la duración y lo recaudado y entre la duración y el score de imdb con valor **0.256352**.

Luego también hay una correlación un poco más baja con valor **0.247049**	 entre el score de imdb y la cantidad de likes de facebook de la película.
